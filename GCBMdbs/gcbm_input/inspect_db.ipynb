```python
import sqlite3
import pandas as pd
```

```python
def show_db(db_name):
    sqliteConnection = sqlite3.connect(db_name)
    cursor = sqliteConnection.cursor()
    
    sql_query = "SELECT name FROM sqlite_master WHERE type='table';"
    cursor.execute(sql_query)
    tables = cursor.fetchall()
    
    for count, table in enumerate(tables, 1):
        print(f'--------------{count}. Table: {table[0]}--------------')

        query = f"SELECT * FROM {table[0]}"
        res = pd.read_sql(query, sqliteConnection)
        print(res)    
```

```python
def compare_dbs(db_name1, db_name2):
    distinct_df = []
    
    sqliteConnection1 = sqlite3.connect(db_name1)
    sqliteConnection2 = sqlite3.connect(db_name2)
    cursor = sqliteConnection1.cursor()
    
    sql_query = "SELECT name FROM sqlite_master WHERE type='table';"
    cursor.execute(sql_query)
    tables = cursor.fetchall()
    
    for count, table in enumerate(tables, 1):
        query = f"SELECT * FROM {table[0]}"
        res1 = pd.read_sql(query, sqliteConnection1)
        res2 = pd.read_sql(query, sqliteConnection2)
        
        if not res1.equals(res2):
            distinct_df.append(f'{count}. {table[0]}')
    
    print(f'Distinct Tables:')
    print(*distinct_df, sep = '\n')
```

# 1. GCBM Rectangle World

```python
db_demo = 'gcbm_input_demorun.db'
show_db(db_demo)
```

# 2. GCBM.Carpathians

```python
db_carpathians = 'gcbm_input_carpathians.db'
show_db(db_carpathians)
```

# 3. Rectange World vs Carpathians diffs

```python
compare_dbs(db_demo, db_carpathians)
```
